# 02 · Modello tempi sul giro (rigoroso)
Confronto tra algoritmi, cross-validation, importanza feature e predetto-vs-reale.

```python
import sys; sys.path.insert(0, '../src')
from f1va import features, models, synthetic
import pandas as pd, numpy as np

# Dataset ampio (stagione sintetica) per un ML significativo. Offline.
laps = synthetic.generate_season(n_races=8, n_drivers=20, seed=0)
X, y = features.build_laptime_dataset(laps)
print(f'{len(y):,} giri, {X.shape[1]} feature')
```

## Confronto modelli (5-fold CV)

```python
models.compare_models(X, y)
```

## Tuning iperparametri (gradient boosting)

```python
from sklearn.model_selection import train_test_split
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2, random_state=0)
m, info = models.tune_laptime_model(Xtr, ytr, n_iter=20, cv=4)
print(info['best_params']); print('CV MAE:', info['cv_mae_s'])
print('Test:', m.evaluate(Xte, yte))
```

## Importanza delle feature

```python
models.permutation_importance_table(m, Xte, yte)
```

## Predetto vs reale

```python
import matplotlib.pyplot as plt
pred = m.predict(Xte)
plt.scatter(yte, pred, s=6, alpha=.3)
lims=[yte.min(), yte.max()]; plt.plot(lims, lims, 'r--')
plt.xlabel('Reale (s)'); plt.ylabel('Predetto (s)')
```
