# 01 · Esplorazione dati F1
Carica una sessione reale via FastF1, pulisce i giri e osserva il degrado gomme.

```python
import sys; sys.path.insert(0, '../src')
from f1va import data as f1data, features, models, strategy
import pandas as pd, numpy as np
```

```python
ses = f1data.load_session(2024, 'Monza', 'R')
raw = f1data.laps_dataframe(ses)
laps = f1data.quicklaps(raw)
print(f'{len(laps)} giri validi, {laps.Driver.nunique()} piloti')
laps.head()
```

## Degrado per mescola

```python
deg = features.degradation_table(laps)
deg
```

```python
import matplotlib.pyplot as plt
for comp, g in laps.groupby(laps.Compound.str.upper()):
    plt.scatter(g.TyreLife, g.laptime_s, s=10, label=comp)
plt.xlabel('Eta gomma (giri)'); plt.ylabel('Tempo (s)'); plt.legend(); plt.title('Degrado gomme')
```
